```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = [11, 7]

data=pd.read_csv("datasets/office_episodes.csv")

print(data.info())
print(data.head())
```

```python
#before filteration
```

```python
plt.scatter(data['episode_number'],data['viewership_mil'])
```

```python
#adding color scheme based on rating value
```

```python
color=[]
for lab,row in data.iterrows():

    if  row['scaled_ratings'] < 0.25:
            color.append("red")
    elif   0.50 > row['scaled_ratings']>= 0.25:
            color.append("orange")
    elif   0.75 > row['scaled_ratings']>= 0.50 : 
            color.append("lightgreen")
    else:
            color.append("darkgreen")
```

```python
#size sys to make difference based on the guest appearances
```

```python
size=[]
for ind,row in data.iterrows():
    if row['has_guests']==False:
        size.append(25)
    else:
        size.append(250)
```

```python
#adding new coloumns to dataset to make difference based on guest appearances
```

```python
data['sizes']=size
data['col']=color
```

```python
#filtering data based on guest appearnces
```

```python
with_guests=data[data['has_guests']==False]
without_guests=data[data['has_guests']==True]
```

```python
# Create a title and axis labels
```

```python
fig=plt.figure()
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.scatter(x=with_guests.episode_number, y=with_guests.viewership_mil, c=with_guests['col'] , s=with_guests['sizes'])
plt.scatter(x=without_guests.episode_number, y=without_guests.viewership_mil, c=without_guests['col'],marker='*', s=without_guests['sizes'])
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.show()
```
